## Preprocessing

```python
from sklearn.model_selection import train_test_split

X = data.drop("prognosis", axis=1)
y = data["prognosis"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=69, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler, RobustScaler
import matplotlib.pyplot as plt

num_cols = ["Age", "Temperature (C)", "Humidity", "Wind Speed (km/h)"]
X_num = data[num_cols]

scalers = {
    'StandardScaler': StandardScaler(),
    'MinMaxScaler': MinMaxScaler(),
    'MaxAbsScaler': MaxAbsScaler(),
    'RobustScaler': RobustScaler()
}

plt.figure(figsize=(18, 10))

for idx, (name, scaler) in enumerate(scalers.items()):
    X_scaled = scaler.fit_transform(X_num)
    plt.subplot(2, 2, idx+1)
    plt.boxplot(X_scaled, labels=num_cols, showfliers=False)
    plt.title(name)
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### The best scaler is StandardScaler and RobustScaler, so I'm gonna use StandardScaler for this data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

## Algorithms

Selection of hyperparameters of models

Optimal parameters were searched for each model (according to cross-validation graphs):

- **KNN:** optimally, k = 2.
- **SVC:** optimal C ≈ 3.
- **Logistic Regression:** optimal C ≈ 100.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt

k_range = range(1, 21)
scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    cv_scores = cross_val_score(knn, X_train, y_train, cv=5, scoring='accuracy')
    scores.append(cv_scores.mean())

plt.plot(k_range, scores, marker='o')
plt.xlabel('Number of neighbors (k)')
plt.ylabel('Average CV accuracy')
plt.title('Selecting k for KNN (Elbow method)')
plt.xticks(k_range)
plt.grid(True)
plt.show()

print(f"Лучшее k: {k_range[np.argmax(scores)]} (accuracy={max(scores):.4f})")
```

```python
C_range = np.logspace(-2, 2, 10)  # значения C от 0.01 до 100
svc_scores = []

for C in C_range:
    svc = SVC(kernel='linear', C=C, random_state=42)
    svc_scores.append(cross_val_score(svc, X_train, y_train, cv=5, scoring='accuracy').mean())

plt.plot(C_range, svc_scores, marker='o')
plt.xscale('log')
plt.xlabel('C (log scale)')
plt.ylabel('Average CV accuracy')
plt.title('Selecting C for SVCs')
plt.grid(True)
plt.show()

print(f"Лучший C: {C_range[np.argmax(svc_scores)]:.2f} (accuracy={max(svc_scores):.4f})")
```

```python
C_range = np.logspace(-2, 2, 10)
logreg_scores = []

for C in C_range:
    logreg = LogisticRegression(C=C, max_iter=1000, multi_class='multinomial', solver='lbfgs')
    logreg_scores.append(cross_val_score(logreg, X_train, y_train, cv=5, scoring='accuracy').mean())

plt.plot(C_range, logreg_scores, marker='o')
plt.xscale('log')
plt.xlabel('C (log scale)')
plt.ylabel('Average CV accuracy')
plt.title('Selecting C for Logistic Regression')
plt.grid(True)
plt.show()

print(f"Лучший C: {C_range[np.argmax(logreg_scores)]:.2f} (accuracy={max(logreg_scores):.4f})")
```

```python
models = {
    'Logistic Regression': LogisticRegression(C=100, max_iter=1000, multi_class='multinomial', solver='lbfgs'),
    'Decision Tree': DecisionTreeClassifier(random_state=69),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=69),
    'Naive Bayes': GaussianNB(),
    'KNN': KNeighborsClassifier(n_neighbors=2),
    'SVC': SVC(kernel='linear', C=3, probability=True, random_state=69)
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(results.keys(), results.values(), color='deepskyblue')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison')
plt.ylim(0.85, 1.0)
plt.xticks(rotation=30, ha='right')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.002,
             f"{bar.get_height():.3f}", ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

- Leaders: SVC, Random Forest
- Logistic Regression and Decision Tree also showed high quality.
- Naive Bayes and KNN lagged behind other models.

## MLPClassifier (Neural Network)

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, random_state=69)
mlp.fit(X_train, y_train)
y_pred = mlp.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print("=== MLPClassifier ===")
print(f"Accuracy: {acc:.4f}")
print(classification_report(y_test, y_pred, target_names=le.classes_))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
results['MLPClassifier'] = acc

# Перестроим barplot
plt.figure(figsize=(10, 6))
bars = plt.bar(results.keys(), results.values(), color='mediumorchid')
plt.ylabel('Accuracy')
plt.title('Comparison of Accuracy models')
plt.ylim(0.85, 1.0)
plt.xticks(rotation=30, ha='right')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.002,
             f"{bar.get_height():.3f}", ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(mlp.loss_curve_)
plt.xlabel("Era")
plt.ylabel("Log-loss (error)")
plt.title("Convergence of learning MLPClassifier")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores = learning_curve(
    mlp, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10)
)

plt.figure(figsize=(8,5))
plt.plot(train_sizes, np.mean(train_scores, axis=1), label='Train')
plt.plot(train_sizes, np.mean(test_scores, axis=1), label='Validation')
plt.xlabel('Train size')
plt.ylabel('Accuracy')
plt.title('Learning curve for MLPClassifier')
plt.legend()
plt.tight_layout()
plt.show()
```

- **Convergence graph (loss curve):** Fast and steady error convergence.
- **Learning curve:** Increased accuracy with an increase in the training sample, lack of retraining.
